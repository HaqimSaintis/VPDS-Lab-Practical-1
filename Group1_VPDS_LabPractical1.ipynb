## Question 2

```python
# Count duplicates
print(f'Total duplicates: {df.duplicated().sum()}')
```

```python
# Show the duplicate rows
duplicate_rows = df[df.duplicated(keep=False)]
duplicate_rows
```

```python
# Display five exact rows that have duplicated values
df[df.duplicated(keep='first')]
```

```python
# Drop duplicates and save to df variable
df = df.drop_duplicates(keep='first', inplace=False).copy()
df
```

```python
# Calculate the age mean
age_mean = round(df['Age'].mean())
print(f'The age mean is {age_mean}')
```

```python
# Fill missing Age with mean value
df['Age'] = df['Age'].fillna(age_mean).astype(int)
df
```

```python
# Find the mode from product category column
product_mode = df['Product_Category'].mode()[0]
print(f'The product mode is "{product_mode}"')
```

```python
# Fill missing product category with mode value
df['Product_Category'] = df['Product_Category'].fillna(product_mode)
df
```

```python
# Drop rows where Customer_Name is missing
df = df.dropna(subset=['Customer_Name'], inplace=False).copy()
df
```

## Question 3

```python
# Calculate the Q1 and Q3
Q1 = df['Total_Spend'].quantile(0.25)
Q3 = df['Total_Spend'].quantile(0.75)
# Display the value of Q1 and Q3
print(f'The value of Q1 is {Q1}')
print(f'The value of Q3 is {Q3}')
```

```python
# Calculate the IQR
IQR = Q3 - Q1
# Display the value of IQR
print(f'The value of IQR is {IQR}')
```

```python
# Define and calculate the bounds
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)
# Display the lower bound and upper bound
print(f'The value of lower bound: {lower_bound}')
print(f'The value of upper bound: {upper_bound}')
```

```python
# Filter the outliers data
df = df[(df['Total_Spend'] >= lower_bound) & (df['Total_Spend'] <= upper_bound)]
df
```

```python
# Print the new shape of the dataset
print(f'''The new shape of the dataset is {df.shape},
where {df.shape[0]} is the number of rows and {df.shape[1]} is the 
number of columns.''')
```

## Question 4

```python
# Function to categorise the total spending
def categorise_total_spend(total_spend):
    if total_spend < 100:
        return 'Low'
    elif 100 <= total_spend <= 300:
        return 'Medium'
    else:
        return 'High'
# Copy the original dataset
df = df.copy()
# Apply the function to the new column in the dataset
df['Spend_Category'] = df['Total_Spend'].apply(categorise_total_spend)
# Display the current modified dataset
df
```

```python
# Standardize the product category text to upper case
df['Product_Category'] = df['Product_Category'].str.upper()
df
```

```python
# Save the fully cleaned and transformed dataset as a new file
df.to_csv('final_P1_data.csv', index=False)
# Display the final dataset
df = pd.read_csv('final_P1_data.csv')
df
```
